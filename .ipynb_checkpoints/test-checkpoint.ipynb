```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('data.csv')
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
df.set_index('ID', inplace=True)
```

```python
df.head(15)
```

```python
df.drop(axis=1, labels=['Flag', 'Photo', 'Club Logo', 'Jersey Number'], inplace=True)
```

```python
df = df[df['Nationality'] == 'Serbia']
```

```python
df_without_categorical_attrs = df._get_numeric_data()
df_num_scaled = pd.DataFrame(MinMaxScaler().fit(df_without_categorical_attrs).transform(df_without_categorical_attrs))
```

```python
df_num_scaled.columns = df_without_categorical_attrs.columns
```

```python
df_num_scaled['ID'] = df.reset_index()['ID']
```

```python
df_num_scaled.set_index('ID').head()
```

```python
est = KMeans(n_clusters=5)
```

```python
df_num_scaled.interpolate(inplace=True)
```

```python
est.fit(df_num_scaled)
```

```python
df['Clusters'] = est.labels_
```

```python
colors = ['red', 'blue', 'green', 'yellow', 'gold']
fig = plt.figure(figsize=(7,7))
for j in range(5):
    print("Klater {}".format(j))
    cluster = df.loc[df['Clusters'] == j].loc[df['Club'] == 'FC Barcelona']
    print(cluster)
    plt.scatter(cluster['Age'], cluster['Overall'], color=colors[j])
```
